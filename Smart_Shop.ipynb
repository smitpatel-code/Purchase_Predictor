```python
import pandas as pd
```

```python
df=pd.read_csv("shop_smart_ecommerce.csv")
df.info()
df.isnull().sum()
```

```python
df.head(70)
```

```python
#Encoding

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le=LabelEncoder()
df["Weekend"]=le.fit_transform(df["Weekend"])
df["Revenue"]=le.fit_transform(df["Revenue"])

ohe = OneHotEncoder(drop="first", sparse_output=False, handle_unknown="ignore")
df["Month"]=ohe.fit_transform(df[["Month"]])
df["VisitorType"]=ohe.fit_transform(df[["VisitorType"]])
```

```python
df.head()
```

```python
#Split Data
X=df.drop(columns=["Revenue"])
X
y=df["Revenue"]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
#Decision_Tree
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
from sklearn.metrics import accuracy_score

acc=accuracy_score(y_test, y_pred)
acc
```

```python
#Pre-Prunning
max_depths = [2, 3, 4, 5, 6, 7, 8, 9, 10, 12, 14, 16, 18, 20]
for depth in max_depths:
    model=tree.DecisionTreeClassifier(max_depth=depth)
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(f"for depth={depth}, accuracy={acc}")
 #Best is depth 4   
```

```python

min_samples_split = [2, 3, 4, 5, 6, 7, 8, 9, 10, 12, 14, 16, 18, 20]
best_acc = 0
best_split = None
for mini in min_samples_split:
    model=tree.DecisionTreeClassifier(max_depth=4,min_samples_split=mini)
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    if acc > best_acc:
        best_acc = acc
        best_split = mini

    print(f"For min_samples_split = {mini}, accuracy = {acc}")

print("Best min_samples_split:", best_split)
print("Best accuracy:", best_acc)
 
```

```python
from sklearn import tree

min_samples_leaf = [2,3,4,5,6,7,8,9,10,12,14,16,18,20]

best_acc = 0
best_leaf = None

for leaf in min_samples_leaf:
    model = tree.DecisionTreeClassifier(
        max_depth=4,
        min_samples_split=2,
        min_samples_leaf=leaf
    )

    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)

    if acc > best_acc:
        best_acc = acc
        best_leaf = leaf

    print(f"For min_samples_leaf = {leaf}, accuracy = {acc}")

print("Best min_samples_leaf:", best_leaf)
print("Best accuracy:", best_acc)
```

```python
#Post Prunning
full_tree = tree.DecisionTreeClassifier(random_state=42)
full_tree.fit(X_train, y_train)
path = full_tree.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas

# train our model for all alphas

trees = []

for alpha in ccp_alphas:
    model = tree.DecisionTreeClassifier(random_state=42, ccp_alpha=alpha)
    model.fit(X_train, y_train)

    trees.append((model, alpha))
best_acc = 0
best_alpha = 0

for model, alpha in trees:
    curr_acc = model.score(X_test, y_test)
    if curr_acc > best_acc:
        best_acc = curr_acc
        best_alpha = alpha

print(best_acc)
```

```python
#Combining all together
best_model = tree.DecisionTreeClassifier(ccp_alpha=best_alpha)
best_model.fit(X_train, y_train)
```

```python
y_pred=best_model.predict(X_test)
acc=accuracy_score(y_test, y_pred)
acc
```
